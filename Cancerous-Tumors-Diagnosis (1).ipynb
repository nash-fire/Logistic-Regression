```python
# import 3 libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dataset for illustration

np.random.seed(42)
```

```python
# Generate 20 samples of benign tumor with random size (0-5) and label 0

benign_samples = pd.DataFrame({
    'size': np.random.rand(20) * 5,
    'label': [0] * 20
})
```

```python
# Generate 20 samples of malignant tumor with random size (5-10) and label 1

malignant_samples = pd.DataFrame({
    'size': 5 + np.random.rand(20) * 5,
    'label': [1] * 20
})
```

```python
# Combine both benign and malignant samples

data = pd.concat([benign_samples, malignant_samples], ignore_index=True)

# Plot the synthetic dataset
plt.figure(figsize=(8, 6))
plt.scatter(data[data['label'] == 0]['size'], data[data['label'] == 0]['label'], color='blue', label='Benign', alpha=0.5, marker='o')
plt.scatter(data[data['label'] == 1]['size'], data[data['label'] == 1]['label'], color='red', label='Malignant', alpha=0.5, marker='x')
plt.xlabel('Tumor Size')
plt.ylabel('Diagnosis (0=Benign, 1=Malignant)')
plt.title('Synthetic Dataset - Tumor Cancer Classification')
plt.legend()
plt.show()
```
